# AI Engineering Fundamentals: Tensors, Distributions, & Randomness

This repository contains exercises and theoretical notes detailing the transition from classical deterministic programming to probabilistic AI engineering. The core focus is on how multi-dimensional arrays (Tensors) and mathematical randomness act as the foundational fuel for Machine Learning and Generative AI.

---

## 1. Randomness & The Seed (PRNG)

Computers are purely logical machines and cannot generate true randomness natively. Instead, they rely on deterministic mathematical formulas known as **Pseudo-Random Number Generators (PRNGs)**.

### The Seed ("The GPS Coordinate")
To begin generating a sequence of random numbers, a PRNG requires an initial state called a **Seed**.
* **Default Behavior:** If no seed is provided, the computer usually defaults to using the current system time in milliseconds as the starting point.
* **The Analogy:** The numerical value of a seed (e.g., `1` or `888`) does not make the output "better" or "more random." Think of the seed simply as a GPS coordinate on an infinite map of numbers. Seed `888` always drops you at the exact same starting location.

### Real-World AI Application: Reproducibility
In a professional machine learning pipeline, you must explicitly set a seed (e.g., `np.random.seed(888)`). This ensures that if another developer clones your repository, their computer will traverse the exact same random path. Without a set seed, a model's initial state is lost forever, making debugging and peer review impossible.

## 2. Probability Distributions

The "shape" of generated random data dictates how a neural network learns. We rely heavily on two primary distributions.

### Uniform Distribution
* **The Concept:** Every single value within a specified range has the exact same probability of being chosen. It is perfectly democratic. For an array of `[1, 2, 3, 4]`, every element has exactly a 25% chance of selection.
* **AI Application:** Used when unbiased selection is mandatory, such as randomly shuffling a dataset or grabbing a random batch of 32 images from a pool of 1,000,000 to train a model.

### Normal (Gaussian) Distribution
* **The Concept:** The classic "Bell Curve." Data is densely clustered around the center and tapers off toward the edges.
* **Mathematical Properties:** It is perfectly symmetrical. The **Mean** (average), **Median** (exact middle), and **Mode** (most frequent) are absolutely equal and sit at the exact peak of the curve. The infinite tails represent real-world anomalies that are rare but mathematically possible.
* **AI Application:** This is critical for **Neural Network Weight Initialization**. 
  * If weights are initialized to exactly `0`, the network fails to break symmetry (millions of neurons will just duplicate the same work).
  * If weights are too large, the math exponentially explodes. 
  * We initialize weights using a Normal distribution centered near `0` (with a tiny standard deviation) to ensure signals flow smoothly through the network without vanishing or exploding.

## 3. Dimensionality & Tensors

In AI engineering, we transition away from standard loops and instead manipulate contiguous blocks of memory shaped into dimensions. These multi-dimensional arrays are called **Tensors**.

* **1D Array (Vector):** A sequence of data points. 
  * *Example:* 100 random temperatures over 100 days.
* **2D Array (Matrix):** A standard grid `(Height, Width)`. 
  * *Example:* A shape of `(1920, 1080)` represents a grayscale image. The array has no "lighting" dimension; the lighting intensity (0 to 255) is the actual number stored inside each cell.
* **3D Array (Tensor):** Adds depth `(Height, Width, Depth)`. 
  * *Example:* A full-color image has the shape `(1920, 1080, 3)`. The `3` represents the Red, Green, and Blue (RGB) color channels.
* **4D Array (Image Batch):** The standard shape for training computer vision models. 
  * *Example:* A tensor shaped `(4, 256, 256, 3)` translates to: *"A batch of 4 distinct images, each sized 256x256, with 3 RGB color channels."* We stack images into 4D batches to feed them into GPUs for massive parallel processing (vectorization).

## 4. Connections to Generative AI

The basic random generation concepts detailed above are the exact mechanics powering state-of-the-art Generative AI.

### Large Language Models (Text Generation)
When predicting the next token, an LLM outputs a **Categorical Distribution** (probabilities mapped across its entire 256,000+ word vocabulary).
* **Temperature:** A hyperparameter that changes how the PRNG samples from this distribution.
  * **Low Temperature:** The model acts deterministically, picking the highest-probability tokens (resulting in concise, consistent text).
  * **High Temperature:** The PRNG is allowed to sample lower-probability tokens, resulting in highly creative, though less predictable, text.

### Diffusion Models (Image Generation)
Models like Stable Diffusion do not generate billions of color pixels out of thin air. They rely directly on PRNGs and Normal Distributions.
1. **The Starting Point:** The model generates a 3D tensor of pure, random **Gaussian noise** (visual static). 
2. **The Denoising Loop:** Over 20 to 50 steps, the AI iteratively subtracts tiny amounts of noise, using your text prompt to sculpt the random static into a coherent image.
3. **The Seed:** Because the starting block of static is generated by an RNG, the seed is everything. If you use the same seed and the same text prompt, the initial static is identical, and the AI will sculpt the exact same image. Changing the seed by even 1 generates a totally unique initial noise block, resulting in a completely different final image.
