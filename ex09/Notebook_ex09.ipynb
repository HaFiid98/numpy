```python
import numpy as np
import itertools as it
TeamCote = np.genfromtxt("model_forecasts.txt", delimiter="," , missing_values=0)
# print(TeamCote)
TeamPermutations = np.array(list(it.permutations(range(10)))).reshape( 3628800 ,5, 2 )
# TeamCote[TeamPermutations]
# print(TeamPermutations[:,1])
# print(TeamPermutations[::,0])
IndexedPermutation = TeamCote[TeamPermutations[:,:,0] , TeamPermutations[:,:,1]  ] 
Squared = IndexedPermutation **2
YScore = np.sum(Squared , axis= 1) 
BestCombination = TeamPermutations[np.argmin(YScore)]
# print(IndexedPermutation[np.argmin(YScore)])

# Transposing here : flip the matrix diagonally columns -> rows ,
# rows -> columns (u will learn more in linear algebra)
print(BestCombination.T)
# print(TeamCote[np.array(list(TeamPermutations))])
```
