```python
import numpy as np 
np.set_printoptions(suppress=True)
# WineData = np.genfromtxt('winequality-red.csv', delimiter=';'  , missing_values= '_' , filling_values= 0 ).astype(np.float32)
WineData = np.genfromtxt('winequality-red.csv', delimiter=';' ,  skip_header=1 , missing_values= ['_' , ''] , filling_values= 0 ).astype(np.float32)
# data size in bytes
# print(WineData)
# DisplayRow =  np.concat( , WineData[6:,])
Indexes = [1 , 6 , 11]
Display = WineData[Indexes].astype(np.float32)
# print(Display)

Display = Display[~np.isnan(Display)]
Display = Display[np.newaxis,:]


#where produce array mask of boolean then match to take only true based on condition 
# return tuple of the newy array with it type
AlcoPerc20 = np.where(WineData[:,10] > 20 )[0].any();
AlcoAvg = np.average(WineData[:,10])
AverageSulf = np.percentile(WineData[:,9] ,20)
mean = np.mean(WineData[:,9])
print("mean: " , mean)
Quality = WineData[WineData[:,9] <= AverageSulf]
# print(np.average(Quality[:,11]))
# Quality = np.percentile(WineData[:,1],20)
maxQuality = WineData[WineData[:,11] == np.max(WineData[:,11])]
lowQuality = WineData[WineData[:,11] == np.min(WineData[:,11])]
best_wines = np.mean(maxQuality,axis=0) 
worst_wines = np.mean(lowQuality,axis=0) 


print("the means of the best Quality and worst quality Wine  Best : " )
print("max quality:\n " , best_wines)
print("low quality:\n " , worst_wines)
print("the average quality score of wines with the 20% least sulphate content." ,np.average(Quality[:,11]))
print("Average Sulfate score over 20% : " ,AverageSulf )
print("Ph 25 percentile : " , np.percentile(WineData[:,8] , 25))
print("Ph 50 percentile : " , np.percentile(WineData[:,8] , 50))
print("Ph 75 percentile : " , np.percentile(WineData[:,8] , 75))
print("Ph mean : " , np.mean(WineData[:,8]))
print("Ph min : " , np.min(WineData[:,8]))
print("Ph max : " , np.max(WineData[:,8]))


print("is There An Wine with Over 20% Alcohol : " , AlcoPerc20 , "\nAverage Alcohol Percentile :" , AlcoAvg)
```
