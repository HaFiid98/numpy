# Python Wheels for AI Engineering

## 1. Overview

When installing a Python package with `pip`, the first thing `pip` tries to find is a **Wheel (`.whl`)**, which is the standard binary distribution format for Python packages.

A wheel is a **pre-built package** that is ready to install. If a compatible wheel exists for your operating system, Python version, and CPU architecture, `pip` simply downloads and installs it.

If no compatible wheel exists, `pip` downloads the package's source code and compiles it locally. This requires build tools such as:

- GCC (Linux)
- Clang (macOS)
- Microsoft Visual C++ (Windows)

---

# 2. Why Wheels Exist

Without wheels:

```
pip install numpy

↓

Download source code

↓

Compile C/C++ code

↓

Link libraries

↓

Install package
```

This process can take several minutes and requires compilers and development libraries.

With wheels:

```
pip install numpy

↓

Download .whl

↓

Copy files into site-packages

↓

Done
```

Installation becomes much faster because the package has already been compiled by the package maintainers.

---

# 3. What Is a Wheel?

A wheel (`.whl`) is **not executable code**.

It is simply a standardized archive (similar to a ZIP file) containing everything needed to install a package.

Typical contents include:

```
numpy/
    __init__.py
    random/
    linalg/
    core/

numpy/core/
    _multiarray_umath.so

METADATA
RECORD
WHEEL
```

Depending on the platform, compiled extension files are:

| Platform | Extension |
|----------|-----------|
| Linux | `.so` |
| Windows | `.pyd` |
| macOS | `.so` or `.dylib` |

---

# 4. How Wheels Are Built

The package maintainer creates the wheel before uploading it to PyPI.

```
Developer writes code

Python
C
C++
Fortran
CUDA

↓

Compiler

↓

Shared libraries

.so
.dll
.pyd

↓

Packaged into

package.whl

↓

Uploaded to PyPI
```

Users normally never compile the package themselves.

---

# 5. Installation Workflow

When you run:

```bash
pip install numpy
```

`pip` performs roughly the following steps:

```
Search PyPI

↓

Compatible wheel found?

        │
   Yes  │  No
        │
        ▼
 Download wheel
        │
 Install files
        │
      Finished

        OR

Download source

↓

Compile using GCC/MSVC/Clang

↓

Build package

↓

Install
```

# 6. Why AI Libraries Use Wheels

Most AI and scientific libraries are written using multiple languages.

```
Python

↓

Calls

↓

C/C++

↓

BLAS / LAPACK

↓

CPU
```

Some libraries also use:

- CUDA
- cuDNN
- oneDNN
- OpenBLAS
- Intel MKL

The Python code is mostly a high-level interface.

The computationally intensive work happens in optimized native code.

Examples include:

- NumPy
- SciPy
- pandas
- scikit-learn
- PyTorch
- TensorFlow
- XGBoost

Almost all of these are distributed as wheels.

---

# 7. Pure Python vs Binary Wheels

## Pure Python Wheel

Contains only Python source files.

```
requests/

↓

.py files only
```

Characteristics:

- No compilation
- Works on every operating system
- One wheel supports all platforms

---

## Binary Wheel

Contains compiled native code.

```
NumPy

↓

.so
.pyd
.dll
```

Characteristics:

- Much faster
- Platform-specific
- Separate wheel for each operating system and architecture

---

# 8. Wheel Filename Explained

Example:

```
numpy-2.3.0-cp313-cp313-manylinux_2_28_x86_64.whl
```

| Part | Meaning |
|------|---------|
| numpy | Package name |
| 2.3.0 | Version |
| cp313 | CPython 3.13 |
| cp313 | ABI compatibility |
| manylinux_2_28 | Linux compatibility tag |
| x86_64 | CPU architecture |

`pip` compares these tags against your current environment to determine whether the wheel is compatible.

---

# 9. Why One Wheel Cannot Run Everywhere

Native binaries differ between operating systems.

| Operating System | Binary Format |
|------------------|---------------|
| Linux | ELF |
| Windows | PE |
| macOS | Mach-O |

Because native binaries are different, a Linux wheel cannot run on Windows, and a Windows wheel cannot run on macOS.

Each platform requires its own compiled wheel.

---

# 10. Wheel vs Java

This comparison is often misunderstood.

| Java | Python |
|-------|---------|
| `.class` | `.so` / `.pyd` |
| `.jar` | `.whl` |

### Java

```
.java

↓

javac

↓

.class

↓

JVM executes bytecode
```

### Python

```
C/C++ source

↓

Compiler

↓

.so / .pyd

↓

Packaged inside

.whl

↓

pip installs

↓

Python imports module
```

A wheel is **not** like Java bytecode.

A wheel is closer to a **JAR file** because both are packaging formats.

The compiled native libraries inside the wheel are roughly analogous to the compiled classes inside a JAR.

---

# 11. Big Picture in AI Engineering

Most AI libraries follow this architecture:

```
Python Code

↓

High-Level API

↓

Native C/C++ Extensions

↓

BLAS / LAPACK / CUDA / cuDNN

↓

CPU / GPU Hardware
```

Python provides the interface, while optimized native libraries perform the heavy numerical computations.

This combination gives Python both ease of use and high performance.

---

# 12. Key Takeaways

- A **wheel (`.whl`)** is Python's standard binary package format.
- Wheels usually contain precompiled native extensions.
- `pip` always tries to install a compatible wheel before compiling source code.
- If no wheel exists, `pip` builds the package locally using a compiler.
- AI libraries such as NumPy, PyTorch, SciPy, TensorFlow, and pandas are almost always distributed as wheels.
- Wheels are packaging formats, **not executable runtimes**.
- A wheel is conceptually closer to a Java **JAR** than to Java bytecode (`.class`).
- Understanding wheels helps explain why Python can efficiently use high-performance C, C++, Fortran, and CUDA code while maintaining a simple programming interface.
